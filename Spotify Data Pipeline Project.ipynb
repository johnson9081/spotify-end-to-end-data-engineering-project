```python
album_list = []
for val in range(len(data['items'])):
    album_dict = {  'album_id': data['items'][val]['track']['album']['id'],
                    'name': data['items'][val]['track']['album']['name'],
                    'release_date': data['items'][val]['track']['album']['release_date'],
                    'total_tracks': data['items'][val]['track']['album']['total_tracks'],
                    'url': data['items'][val]['track']['album']['external_urls']['spotify']}
    album_list.append(album_dict)
```

```python
artist_list = []
for val in range(len(data['items'])):
    for artist in range(len(data['items'][val]['track']['artists'])):
        artist_dict = {'artist_id': data['items'][val]['track']['artists'][artist]['id'],
                     'artist_name': data['items'][val]['track']['artists'][artist]['name'],
                     'external_url': data['items'][val]['track']['artists'][artist]['href']}
        artist_list.append(artist_dict)
```

```python
artist_list
```

```python
songs_list = []
for val in data['items']:
    song_id = val['track']['id']
    song_name = val['track']['name']
    song_duration = val['track']['duration_ms']
    song_url = val['track']['external_urls']['spotify']
    song_popularity = val['track']['popularity']
    song_added = val['added_at']
    album_id = val['track']['album']['id']
    artist_id = val['track']['album']['artists'][0]['id']
    song_element = {'song_id': song_id,'song_name': song_name, 'duration_ms': song_duration, 'url': song_url, 'popularity': song_popularity,
                   'song_added': song_added, 'album_id': album_id, 'artist_id': artist_id}
    songs_list.append(song_element)
```
